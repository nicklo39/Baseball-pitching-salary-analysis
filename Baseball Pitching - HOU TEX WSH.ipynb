```python
import pandas as pd
import seaborn as sns
%matplotlib inline 
```

```python
#Read in the data frame and take a quick peek to see if there's missing info

df = pd.read_csv('pitchingsample.csv')
df.info()
```

```python
#Split the data into the respective teams 

dfHOU = (df['Teams'] == 'HOU')
HOUSTON = df[dfHOU]

dfTEX = (df['Teams'] == 'TEX')
TEXAS = df[dfTEX]

dfWSH=(df['Teams']=='WSH')
WASHINGTON = df[dfWSH]

dfWSH2=(df['Teams']=='WSH')
WASHINGTON2 = df[dfWSH2]
WASHINGTON2.drop([60], inplace = True) #Scherzer being removed for skewed data - visualize with both
```

```python
sns.jointplot(x='Innings', y = 'Contract', data=HOUSTON, kind='kde', color='orange')
sns.jointplot(x='Innings', y = 'Contract', data=TEXAS, kind='kde')
sns.jointplot(x='Innings', y = 'Contract', data=WASHINGTON, kind='kde', color='red')
sns.jointplot(x='Innings', y = 'Contract', data=WASHINGTON2, kind='kde', color='red')
```
